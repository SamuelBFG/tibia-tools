```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get('https://guildstats.eu/bosses?monsterName=&world=Dibra&rook=0')
if req.status_code == 200:
    print('Requisição bem sucedida!')
    content = req.content
```

```python
soup = BeautifulSoup(content, 'html.parser')
table = soup.find(name='table')
```

```python
df = pd.read_html(str(table))[0]
```

```python
df.columns = df.columns.get_level_values(0)
df.columns = ['#', 'Boss name', 'Image', 'Killed Bosses (Yesterday)', 'Killed Players (Yesterday)',
              'Killed Bosses (All Time)', 'Killed Players (All Time)', 'Last Seen', 'Introduced']
df[:10]
```

```python
df = df.drop(['Image', 'Killed Players (Yesterday)', 'Killed Bosses (All Time)', 'Killed Players (All Time)',
        'Introduced'], axis=1)
df[:10]
```
